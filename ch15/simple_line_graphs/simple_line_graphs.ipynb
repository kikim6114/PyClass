### 기본 설정  

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns  # 스타일 설정 "plt.style.use('seaborn-v0_8')" 을 사용하는 경우에 필요하다.
```

#### 그래프에 사용될 글꼴의 크기 정의 및 한글 사용 설정

```python
plt.rcParams['font.size'] = 14
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['legend.fontsize'] = 14
plt.rcParams['xtick.labelsize'] = 8
plt.rcParams['ytick.labelsize'] = 8
plt.rcParams['axes.labelsize'] = 10
plt.rcParams["font.family"] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

`save_fig()`: 생성되는 그래프를 `images` 폴더에 저장하는 함수

```python
from pathlib import Path

IMAGES_PATH = Path() / "images"
IMAGES_PATH.mkdir(parents=True, exist_ok=True)

def save_fig(fig_id, tight_layout=True, fig_extension="png", resolution=300):
    path = IMAGES_PATH / f"{fig_id}.{fig_extension}"
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)
```

#### 단순 선 그래프

```python
squares = [1, 4, 9, 16, 25]

fig, ax = plt.subplots()
ax.plot(squares)
save_fig("mpl_squares_0_first_version")
plt.show()
```

#### 레이블 타입과 선 두께 변경

```python
fig, ax = plt.subplots()
ax.plot(squares, linewidth=3)

# 그래프 제목 및 레이블 설정.
ax.set_title("수와 제곱값", fontsize=24)
ax.set_xlabel("값", fontsize=14)
ax.set_ylabel("제곱값", fontsize=14)

# tick 크기 설정.
ax.tick_params(labelsize=14)

save_fig("mpl_squares_1_labeled")
plt.show()
```

#### Axes vs. Axis

```python
X = np.linspace(-np.pi, np.pi, 256, endpoint=True) # X: [-π, +π] 의 256개 값으로 된 numpy array
C,S = np.cos(X), np.sin(X)
Y = [np.cos(X), np.sin(X)]
fig, axs = plt.subplots(2, 1)
axs[0].plot(X, Y[0])
axs[1].plot(X, Y[1])

save_fig("subplots(2, 1)")
plt.show()
```

#### 그래프 정정하기

```python
input_values = [1, 2, 3, 4, 5]  # x 값
squares = [1, 4, 9, 16, 25]     # y 값

fig, ax = plt.subplots()
ax.plot(input_values, squares, linewidth=3)   # ax.plot(독립변수, 종속변수, 옵션들)

# 그림의 제목과 레이블의 서체 크기 지정.
ax.set_title("수와 제곱값", fontsize=24)
ax.set_xlabel("값", fontsize=14)
ax.set_ylabel("제곱값", fontsize=14)

# Tick 레이블의 서체 크기 지정.
ax.tick_params(labelsize=14)

save_fig("mpl_squares_2_corrected_plot")
plt.show()
```

#### `scatter()`로 포인트 열 그리기

```python
x_values = [1, 2, 3, 4, 5]
y_values = [1, 4, 9, 16, 25]

plt.style.use('seaborn-v0_8')
sns.set_theme(font="Malgun Gothic", rc={"axes.unicode_minus":False})  # 반드시 plt.style.use('seaborn-v0_8') 다음에 사용

fig, ax = plt.subplots()
ax.scatter(x_values, y_values, s=100)

# 그래프 제목 및 레이블 설정.
ax.set_title("수와 제곱값", fontsize=24)
ax.set_xlabel("값", fontsize=14)
ax.set_ylabel("제곱값", fontsize=14)

# Tick 레이블의 서체 크기 지정.
ax.tick_params(labelsize=14)

save_fig("scatter_squares_2_series_of_points")
plt.show()
```

#### 데이터 자동 생성

```python
x_values = range(1, 1001)
y_values = [x**2 for x in x_values]

plt.style.use('seaborn-v0_8')
sns.set_theme(font="Malgun Gothic", rc={"axes.unicode_minus":False})  # 반드시 plt.style.use('seaborn-v0_8') 다음에 사용

fig, ax = plt.subplots()
ax.scatter(x_values, y_values, s=10)

# 그래프 제목 및 레이블 설정.
ax.set_title("수와 제곱값", fontsize=24)
ax.set_xlabel("값", fontsize=14)
ax.set_ylabel("제곱값", fontsize=14)

# Tick 레이블의 서체 크기 지정.
ax.tick_params(labelsize=14)

# 각 축의 범위 설정.
ax.axis([0, 1100, 0, 1_100_000])

save_fig("scatter_squares_3_calculating_automatically")
plt.show()
```

#### `Colormap` 사용하기

```python
x_values = range(1, 1001)
y_values = [x**2 for x in x_values]

plt.style.use('seaborn-v0_8')
sns.set_theme(font="Malgun Gothic", rc={"axes.unicode_minus":False})  # 반드시 plt.style.use('seaborn-v0_8') 다음에 사용

fig, ax = plt.subplots()
ax.scatter(x_values, y_values, c=y_values, cmap=plt.cm.Blues, s=10)

# 그래프 제목 및 레이블 설정.
ax.set_title("수와 제곱값", fontsize=24)
ax.set_xlabel("값", fontsize=14)
ax.set_ylabel("제곱값", fontsize=14)

# Tick 레이블의 서체 크기 지정.
ax.tick_params(labelsize=14)

# 각 축의 범위 설정.
ax.axis([0, 1100, 0, 1_100_000])
ax.ticklabel_format(style='plain')

save_fig("scatter_squares_5_colormap")
plt.show()
```
